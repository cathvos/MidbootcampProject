Based on the boxplots and the histograms, we made the following consideration:
- view: dummify in 1 with view and 0 no view 
- condition: group 1 and 2, and dummify 
- sqft_basement: dummify in 1 with basement and 0 no basement 
- sqft_lot15: drop the 3 cases for which sqft_lot15 is > 500000

```python
def dummify(x): 
    if x == 0:
        return 0
    else:
        return 1
```

```python
df_m2['view_dummy']=list(map(dummify,df_m2['view'])) # dummify view
```

```python
df_m2.drop(['view'], axis = 1, inplace=True) #drop original column
```

```python
df_m2['condition'] = df_m2['condition'].replace(1,2) #grouping condition 1 and 2 in one category
```

```python
df_dummy=pd.get_dummies(df_m2['condition']) #creating dummy condition
```

```python
df_m2=pd.concat([df_m2,df_dummy],axis=1)
```

```python
df_m2.rename(columns={2: 'cond2', 3: 'cond3', 4: 'cond4', 5: 'cond5'},inplace=True, errors='raise')
```

```python
df_m2.drop(['condition'], axis = 1, inplace=True) #drop original column
```

```python
df_m2['basement_dummy']=list(map(dummify,df_m2['sqft_basement'])) #dummify basement
```
